```python
# Calculate percentage damage per section

mask = csv['Structure'] == 'Total atlas' # The region here is total atlas
total_atlas = csv[mask]
total_atlas
```

```python
tot_and_sections = total_atlas.iloc[:,3:]
tot_and_sections
```

```python
non_app = tot_and_sections.iloc[0,:]
damaged = tot_and_sections.iloc[-1,:]
total_counts = non_app + damaged
perc_damage = damaged / total_counts
```
