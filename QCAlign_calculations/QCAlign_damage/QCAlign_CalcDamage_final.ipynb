## Task 1: Calculate percentage damage per section

```python
# Identify part of the df to work with. 
# assign all rows with "total atlas" in "Structure" column to total_atlas variable.

mask = csv['Structure'] == 'Total atlas' # The region here is total atlas
total_atlas = csv[mask]
total_atlas
```

```python
# locate columns that contain info about the series (total) or the sections

tot_and_sections = total_atlas.iloc[:,3:]
tot_and_sections
```

```python
non_app = tot_and_sections.iloc[0,:] # pandas series for count of N/A markers per section
damaged = tot_and_sections.iloc[-1,:] # pandas series for damage markers per section
total_counts = non_app + damaged # pandas series for total markers per section
perc_damage = damaged / total_counts # pandas series for percentage damage per section
```

```python
perc_damage
```

```python
# plot panda.series
plt = perc_damage.plot.bar(x="section", y = "% damage", color=['b'])
```

```python
# calculate total counts per region

# list comprehension

total = [x + y for x,y in zip(region_na, region_damage)]
total
```

```python
# compare lists

compare_lists = zip(regions, region_damage, total)
zipped_lists = list(compare_lists)
print(zipped_lists)
```

```python
#perc_region_damage = [x/y for x,y in zip(region_damage, total)]
#This does not work because total is sometimes 0.

perc_region_damage = []
for x,y in zip(region_damage, total):
        if y!=0:
            perc_region_damage.append(x/y)
        else:
            perc_region_damage.append(0)
```

```python
perc_region_damage
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(18,5))
ax.bar(regions, perc_region_damage)

ax.set_ylabel('Damage (%)')
ax.set_title('Percentage damage per atlas region')
plt.xticks(rotation = "vertical",fontsize = "x-small")

plt.savefig("Percentage_damage_per_region.svg", bbox_inches = "tight")
plt.show()
```
