## Task 1: Calculate percentage damage per section

```python
# Create a mask that identifies results where "Structure == "Total atlas". 
# Apply the mask to the df and assign result to the total_atlas variable.

mask = df['Structure'] == 'Total atlas' # The region here is total atlas. Can also do total > 1000
print(mask)
total_atlas = df[mask] # assign this mask to total_atlas variable
total_atlas
```

```python
#Select data in dataframe using .iloc[row selection][column selection]

non_app = total_atlas.iloc[0,3:]  # pandas series for count of N/A markers per section. 
#print(non_app)
damaged = total_atlas.iloc[-1,3:] # pandas series for damage markers per section
#print(damaged)
total_counts = non_app + damaged # calculate total number of markers per section
#print(total_counts)
perc_damage = damaged / total_counts # calculate percentage damage per section
#print(perc_damage)
```

```python
# plot panda.series
sectionplot = perc_damage.plot.bar(xlabel="section", ylabel = "damage", color=['b'])
```

### Use .iloc or .loc to select data from a dataframe

```python
# Select data corresponding to no. of N/A markers per region. 

#Using iloc[row,column]. This means every 5th row starting with first one, 4rd column)
#region_non_app = df.iloc[::5,3]

#But, an alternative / more robust method.
region_nan_mask = df['Status'].isna() # is not a number
region_na = df.loc[region_nan_mask, 'Total'] # .loc[row,column] selection using label name.
#print(region_na)

#region_total_count = region_non_app + region_damaged
#region_perc_damage = region_damaged / region_total_count

region_name = df.loc[region_nan_mask,'Structure'] # locate the region names
region_name
```

```python
# This is a method for calc percentage damage per region based on two lists. Lists do not support division by 0. But numpy.array allows this. 

#perc_region_damage = []
#for x,y in zip(region_damage, total):
#        if y!=0:
#            perc_region_damage.append(x/y)
#        else:
#            perc_region_damage.append(0)
```

```python
perc_region_damage = np.divide(region_damaged.values, total) # use numpy divide function to calculate percentage damage per region. 
perc_region_damage = np.nan_to_num(perc_region_damage, nan = 0) #cannot divide by zero, therefore replace nan with 0.
print(perc_region_damage)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(18,5))
ax.bar(region_name, perc_region_damage)

ax.set_ylabel('Damage')
ax.set_title('Percentage damage per atlas region')
plt.xticks(rotation = "vertical",fontsize = "x-small")

plt.savefig("Percentage_damage_per_region.svg", bbox_inches = "tight")
plt.show()
```
