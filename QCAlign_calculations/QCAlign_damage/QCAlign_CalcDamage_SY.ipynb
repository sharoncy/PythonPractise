```python
mask = csv['Structure'] == 'Isocortex' # The region here is isocortex
region = csv[mask]
region
```

iloc and loc are specific to Pandas and let you quickly select rows and columns in DataFrames

.iloc: selecting data by row numbers 
.loc: selecting data by label or a conditional statement
Each column is a variable and is usually named. 

data.iloc[row selection],[column selection]



```python
#This is all for isocortex only


region_sections = region.iloc[:,3:] # fourth and above columns of the dataframe
not_applicable =  region_sections.iloc[0,:] # 
damaged = region_sections.iloc[-1,:]
total_counts = not_applicable + damaged
percent_damaged = damaged / total_counts
print(f"the damaged percent is {percent_damaged * 100}")
```
